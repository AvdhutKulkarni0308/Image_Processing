```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_histogram_gray(img, title):
    plt.figure(figsize=(5,4))
    plt.hist(img.flatten(), 256, [0,256])
    plt.title(title)
    plt.xlabel("Gray level")
    plt.ylabel("Frequency")
    plt.show()
```

```python
img_color = cv2.imread('input.jpg')   # BGR image from OpenCV
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

plt.imshow(img_color)
plt.title("Original Color Image")
plt.axis('off')
```

```python
plot_hist(img_color, "Histogram  Original  Image")
```

```python
img = cv2.imread('input.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img = cv2.imread('input.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
```

```python
plot_hist(img, "Histogram  Original Grayscale Image")
```

```python
negative = 255 - img

plt.imshow(negative, cmap='gray')
plt.title("Image Negative")
plt.axis('off')
```

```python
plot_hist(negative, "Histogram  Image Negative")
```

```python
rmin = np.min(img)
rmax = np.max(img)

stretched = ((img - rmin) / (rmax - rmin)) * 255
stretched = stretched.astype(np.uint8)

plt.imshow(stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')
```

```python
plot_hist(stretched, "Histogram  Contrast Stretched Image")
```

```python
bit_planes = []

for i in range(8):
    plane = (img >> i) & 1
    bit_planes.append(plane)

plt.figure(figsize=(10,4))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(bit_planes[i], cmap='gray')
    plt.title(f'Bit {i}')
    plt.axis('off')
plt.show()

```

```python
lower = 100
upper = 180

sliced_img = img.copy()
mask = (img >= lower) & (img <= upper)
sliced_img[mask] = 255

plt.imshow(sliced_img, cmap='gray')
plt.title("Gray Level Slicing")
plt.axis('off')
```
